```python
import numpy as np
# from google.colab.patches import cv2_imshow
import cv2

def grayscale_python(im_in, H):
    """
    Applies a linear color transformation to an image, similar to the provided MATLAB code.

    Args:
        im_in (np.ndarray): The input image, expected to be a 3D NumPy array
                             with shape (height, width, channels).
        H (np.ndarray): The transformation matrix. Its shape should be
                        (output_channels, input_channels). For grayscale,
                        output_channels would typically be 1.

    Returns:
        np.ndarray: The transformed image.
    """

    y_res, x_res, input_channels = im_in.shape
    im_out = np.zeros((y_res, x_res, input_channels), dtype=im_in.dtype)

    for i in range(y_res):
        for j in range(x_res):
            x_vector = im_in[i, j, :]
            y_transformed = H @ x_vector
            im_out[i, j, :] = y_transformed

    return im_out

if __name__ == '__main__':
    grayscale_H = np.array([[0.2989, 0.5870, 0.1140]], dtype=np.float32)
    H = np.array([[0.2126, 0.7152, 0.0722], [-0.09991, -0.33609, 0.436], [0.615, -0.55861, -0.05639]])
    image = cv2.imread('obraz.bmp')
    image_float = image.astype(np.float32) / 255.0
    # cv2_imshow(image)
    cv2.imshow('Obraz oryginalny', image)

    grayscale_image = grayscale_python(image_float, grayscale_H)
    # show the image
    print("Obraz w formacie YUV:")
    # cv2_imshow(grayscale_image*255)
    cv2.imshow('Obraz w formacie YUV', grayscale_image)

    
    grayscale_image_y = grayscale_image[:,:,0]
    grayscale_image_u = grayscale_image[:,:,1]
    grayscale_image_v = grayscale_image[:,:,2]

    #normalizacja kanalow u v
    grayscale_image_u = (grayscale_image_u - np.min(grayscale_image_u)) / (np.max(grayscale_image_u) - np.min(grayscale_image_u))
    grayscale_image_v = (grayscale_image_v - np.min(grayscale_image_v)) / (np.max(grayscale_image_v) - np.min(grayscale_image_v))

    print("Kanał Y:")
    # cv2_imshow(grayscale_image_y*255)
    cv2.imshow('Kanał Y', grayscale_image_y*255)
    print("Kanał U:")
    # cv2_imshow(grayscale_image_u*255)
    cv2.imshow('Kanał U', grayscale_image_u*255)
    print("Kanał V:")
    # cv2_imshow(grayscale_image_v*255)
    cv2.imshow('Kanał V', grayscale_image_v*255)

    print("YUV -> RGB")
    # cv2_imshow(grayscale_python(grayscale_image,np.flip(H))*255)
    cv2.imshow('YUV -> RGB', grayscale_python(grayscale_image,np.flip(H))*255)


    # KONWERSJA RGB -> YCBR
    kr, kg, kb = 0.299, 0.587, 0.114
    H_ycbr = np.array([[kr,kg,kb],[-0.5*kr/(1-kb), -0.5*kg/(1-kb),0.5], [0.5,-0.5*kg/(1-kr), -0.5*kb/(1-kr)]])

    grayscale_image_ycbr = grayscale_python(image_float, H_ycbr)
    print("obraz w formacie YCBR:")
    # cv2_imshow(grayscale_image_ycbr*255)
    cv2.imshow('Obraz w formacie YCBR', grayscale_image_ycbr*255)


    print("Kanał ")
```
